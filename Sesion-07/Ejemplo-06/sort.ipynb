### 2. Desarrollo:

```python
import pandas as pd
```

```python
df = pd.read_csv('../../Datasets/new_york_times_bestsellers-dirty.csv', index_col=0)
df.head(3)
```

Por ejemplo, queremos ordenar nuestras entradas empezando por el libro de mayor precio hasta el libro de menor precio. Esto es un ordenamiento usando la columna 'price.numberDouble' de manera descendente. Esto se haría de la siguiente manera usando el método `sort_values()`:

```python
df.sort_values("price.numberDouble", ascending=False).head(5)
```

Si ordenamos en base a 'published_date.numberLong' podemos también obtener la lista desde la publicación más antigua hasta la publicación más reciente:

```python
ordenado_por_fecha = df.sort_values("published_date.numberLong")
ordenado_por_fecha
```

Per como seguramente no lo crees aún, vamos a convertir la fechas timestamp a datetime con el método `to_datetime()`:

```python
ordenado_por_fecha["published_date.numberLong"] = \
    pd.to_datetime(ordenado_por_fecha["published_date.numberLong"],
    unit="ms")
ordenado_por_fecha[["amazon_product_url",
    "published_date.numberLong"]]
```

Si convertimos 'published_date.numberLong' a `datetime` desde el inicio, podemos también ordenar usando el tipo datetime:

```python
df['published_date.numberLong'] = pd.to_datetime(
    df['published_date.numberLong'], unit='ms')
```

```python
df.sort_values('published_date.numberLong').head(3)
```

Por ejemplo, podríamos primero filtrar para sólo tener los libros de la editorial que tiene más libros como 'best sellers' y después ordenarlos del que pasó más días en la lista de 'best sellers' al que pasó menos días en la lista.

Primero encontramos la editorial con más bestsellers con el método `value_counts()`:

```python
# encuentra la editorial
df["publisher"].value_counts()
```

Filtra sólo los que perteneces a la editorial encontrada:

```python
df_1 =  df[ df["publisher"] == "Putnam" ]
df_1.head(3)
```

Ordenando en base a la cantidad de días como bestsellers:

```python
# ordena aquí
df_1.sort_values("weeks_on_list.numberInt", ascending=False)
```
